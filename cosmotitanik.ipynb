```python
df = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
df
```

```python
#Удаляем строки где есть пропуски
df.loc[df['VIP'] == '[null]', 'VIP'] = np.nan
df.loc[df['Cabin'] == '[null]', 'Cabin'] = np.nan
df = df.dropna(axis=0)
df
```

```python
df.describe()
```

```python
lenc_PassengerId = preprocessing.LabelEncoder() #создаем функцию превращаю категориальные данные в числовые
lenc_PassengerId.fit(df['PassengerId']) #обучаем ее на нашем столбце 
df['PassengerId'] = lenc_PassengerId.transform(df['PassengerId'])
```

```python
lenc_hland = preprocessing.LabelEncoder() #создаем функцию превращаю категориальные данные в числовые
lenc_hland.fit(df['HomePlanet']) #обучаем ее на нашем столбце 
df['HomePlanet'] = lenc_hland.transform(df['HomePlanet'])
```

```python
lenc_CryoSleep = preprocessing.LabelEncoder() #создаем функцию превращаю категориальные данные в числовые
lenc_CryoSleep.fit(df['CryoSleep']) #обучаем ее на нашем столбце 
df['CryoSleep'] = lenc_CryoSleep.transform(df['CryoSleep'])
```

```python
lenc_Destination = preprocessing.LabelEncoder() #создаем функцию превращаю категориальные данные в числовые
lenc_Destination.fit(df['Destination']) #обучаем ее на нашем столбце 
df['Destination'] = lenc_Destination.transform(df['Destination'])
```

```python
lenc_VIP = preprocessing.LabelEncoder() #создаем функцию превращаю категориальные данные в числовые
lenc_VIP.fit(df['VIP']) #обучаем ее на нашем столбце 
df['VIP'] = lenc_VIP.transform(df['VIP'])
```

```python
lenc_Transported = preprocessing.LabelEncoder() #создаем функцию превращаю категориальные данные в числовые
lenc_Transported.fit(df['Transported']) #обучаем ее на нашем столбце 
df['Transported'] = lenc_Transported.transform(df['Transported'])
df
```

```python
in_colnames = ['HomePlanet','CryoSleep',
               'VIP', 'RoomService',
              'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
out_colnames = ['Transported']
```

```python
df_in = df[in_colnames]
df_out = df[out_colnames]
```

```python
inputs_train, inputs_test, expected_output_train, expected_output_test = train_test_split(df_in, df_out, test_size = 0.33, random_state = 42)
```

```python
print(inputs_train.head())
print(expected_output_train.head())
```

```python
rf = RandomForestClassifier(n_estimators=55)
rf.fit(inputs_train, expected_output_train)
```

```python
accuracy = rf.score(inputs_test, expected_output_test)
print("Accuracy = {}%".format(accuracy * 100))
```

```python
#функция поиска пропущенный значений
def missing_values(df):
    missing_percent = df.isnull().sum() * 100 / df.shape[0]
    df_missing_percent = pd.DataFrame(missing_percent).round(2)
    df_missing_percent = df_missing_percent.reset_index().rename(
        columns={"index": "Feature", 0: "Missing Percentage (%)"}
    )
    df_missing_value = df.isnull().sum()
    df_missing_value = df_missing_value.reset_index().rename(
        columns={"index": "Feature", 0: "Missing Values"}
    )

    final = df_missing_value.merge(
        df_missing_percent, how="inner", left_on="Feature", right_on="Feature"
    )
    final = final.sort_values(by="Missing Percentage (%)", ascending=False)
    return final
```
